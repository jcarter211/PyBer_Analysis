```python
%matplotlib inline
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
# Load in csv
pyber_ride_df = pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
```

```python
# Set x-axis and tick locations.
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
# Plot the data.
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```

```python
pyber_ride_df.plot.bar(x="Month", y="Avg. Fare ($USD)")
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(pyber_ride_df["Avg. Fare ($USD)"])
stdev
```

```python
#Add error bars with caps 
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=stdev, capsize=3)
```

```python
#Change the color of the bars to skyblue
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=stdev, capsize=3, color="skyblue")
```

```python
#Rotate the labels on the x-axis to be horizontal
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=stdev, capsize=3, color="skyblue")
plt.xticks(rotation=0)
```

```python
#Set the y-axis increment to every $5
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", kind='bar', yerr=stdev, capsize=3, color="skyblue")
plt.xticks(rotation=0)
plt.yticks(np.arange(0,55, step=5.0))
```
